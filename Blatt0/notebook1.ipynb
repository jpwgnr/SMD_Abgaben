Es soll ein Histogramm zu den Daten aus der Datei `Groesse_Gewicht.txt` erzeugt werden und bestimmt werden, welche Unterschiede je nach Anzahl der Bins zu erkennen sind und welche Bins vern√ºnftig erscheinen und weshalb. 

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
gewicht, groesse= np.genfromtxt("data.txt", unpack=True)
```

```python
def subplot(x):
    a=[5, 10, 15, 20, 30, 50]
    for j,i in enumerate(a):
        plt.subplot(3,2,j+1)
        plt.hist(x, i)
        plt.tight_layout()
```

```python
plt.figure(1)
subplot(gewicht)
plt.figure(2)
subplot(groesse)
```

Als Unterschiede sind dabei zu erkennen, dass die ersten beiden Plots sehr ungenau und die letzten drei Plots oftmals etwas zu detailiert wirken. Eine ziemlich gute Aussage lassen bei beiden Messungen die dritten Plots treffen.
