### Nr. 12a)

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

Lesen der Daten

```python
p0 = pd.read_hdf('zwei_populationen.h5', key="P_0_10000", mode="r")
p1 = pd.read_hdf('zwei_populationen.h5', key="P_1", mode="r")
p0 = p0.values
p1 = p1.values
```

Berechne Mittelwerte $\vec{\mu_\text{P0}}$ und $\vec{\mu_\text{P1}}$.

```python
mup0 = np.mean(p0, axis=0)
mup1 = np.mean(p1, axis=0)
```

### b)
Berechne Kovarianzmatrix $V_\text{P0}$ und $V_\text{P1}$ der beiden Populationen, sowie die kombinierte Kovarianzmatrix $V_\text{P0,P1}$.

```python
covp0 = np.cov(p0.T)
covp1 = np.cov(p1.T)
covp01 = covp0+covp1
```

### c)
Konstruiere Fisher Diskriminante $\vec{\lambda} = \lambda \cdot \vec{e}_{\vec{\lambda}}$. 

```python
lambda_min = np.linalg.inv(covp01)@(mup0-mup1)
print(f"Die Geradengleichung ist eine Achse durch den Nullpunkt mit dem Richtungsvektor({lambda_min[0]:.2f}, {lambda_min[1]:.2f}).")
```

```python
signal = p0@lambda_min
untergrund = p1@lambda_min
plt.hist(signal, bins=50, density=True, label="Signal")
plt.hist(untergrund, bins=50, density=True, label="Untergrund")
plt.legend()
None
```

```python
def effizienzSig(lamb, Sig):
    eff = []
    for i in range(len(lamb)):
        eff = np.append(eff, len(Sig[Sig>lamb[i]])/len(Sig)) 
    return eff

def reinheitSig(lamb, Sig, Unt):
    rein = []
    for i in range(len(lamb)):
        rein = np.append(rein, len(Sig[Sig>lamb[i]])/(len(Sig[Sig>lamb[i]])+ len(Unt[Unt>lamb[i]])))
    return rein
                                           
lambda_eff = np.linspace(np.min([np.min(signal), np.min(untergrund)]), np.max([np.max(signal), np.max(untergrund)])-0.1, 10000)
reinp0dim1 = reinheitSig(lambda_eff, signal, untergrund)
effp0dim1 = effizienzSig(lambda_eff, signal)

                                           
plt.plot(lambda_eff, effp0dim1, label="Effizienz")
plt.plot(lambda_eff, reinp0dim1, label="Reinheit")
plt.legend()
None
```

### f)

```python
def SigToBack(lamb, Sig, Unt):
    sigtoback = []
    for i in range(len(lamb)):
        sigtoback = np.append(sigtoback, len(Sig[Sig>lamb[i]])/(len(Unt[Unt>lamb[i]])))
    return sigtoback

def Signifikanz(lamb, Sig, Unt):
    sign = []
    for i in range(len(lamb)):
        sign = np.append(sign, len(Sig[Sig>lamb[i]])/np.sqrt((len(Unt[Unt>lamb[i]]))+len(Sig[Sig>lamb[i]])))
    return sign

lambda_sigzuUnt = np.linspace(np.min([np.min(signal), np.min(untergrund)]), np.max(np.max(untergrund))-0.1, 10000)
lambda_sign = np.linspace(np.min([np.min(signal), np.min(untergrund)]), np.max([np.max(signal), np.max(untergrund)])-0.1, 10000)
SigToBack = SigToBack(lambda_sigzuUnt, signal, untergrund)
Signifikanz= Signifikanz(lambda_sign, signal, untergrund)

plt.figure(1)
plt.plot(lambda_sigzuUnt, SigToBack, label="Verhltnis Signal/Background")
plt.legend()
plt.figure(2)
plt.plot(lambda_sign, Signifikanz, label="Signifikanz")
plt.legend()


None
```

# b) 

### Beschreibung der Funktionsweise

Die Hauptkomponentenanalyse sucht nach einer Basis im Raum indem die Varianz entlang der Basisvektoren maximiert wird.

Gegeben seien also  $N$  Datenpunkte mit $d$  Dimensionen die auf  $k \, <\, d$  Dimensionen transformiert werden sollen.

Dazu wird der Raum in eine neue Basis transformiert. Es werden also eventuell mehrere Dimensionen/Attribute zu einer neuen Zusammengefasst.

### Grober Ablauf der PCA

* Zentriere die Daten auf ihren Mittelwert.
* Berechne die Kovarianzmatrix aus der Datenmatrix  $\mathbf{X}$.
* Berechne Eigenwerte und Eigenvektoren der Matrix.
* Whle die $k$ grten Eigenwerte und zugehrigen Eigenvektoren aus.
* Bilde eine  $d \times k$  Matrix  $\mathbf{W}$  mit den $k$ Eigenvektoren als Spalten.
* Wende  $\mathbf{W}$ auf jede Zeile aus $x$ auf $\mathbf{X}$ an, $ x=\mathbf{W}^T \cdot x^T$. 


```python
pca = PCA(n_components=4)
pca.fit(X)
covX = pca.get_covariance()
l, W = np.linalg.eigh(covX)
l = l[::-1]
W = W[:, ::-1]
pca2 = PCA(n_components=4)
transformed = pca2.fit_transform(X)
plt.figure(10)
plt.hist(transformed[:,0])
plt.figure(11)
plt.hist(transformed[:,1])
plt.figure(12)
plt.scatter(transformed[:, 0], transformed[:,1], c=labels_true, s=50, cmap=discrete_cmap)
#plt.scatter(transformed[:, 2], transformed[:,3], c=labels_true, s=50, cmap=discrete_cmap)
None
```
